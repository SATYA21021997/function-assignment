1. What is the difference between a function and a method in Python?

ANS - A function is a block of code that performs a specific task. It can be defined using the def keyword. It can be called independently and is not bound to any specific object or class.
A method is a function that is associated with an object or class. It is typically defined inside a class and is meant to operate on instances of that class or the class itself.

2.  Explain the concept of function arguments and parameters in Python.

ANS - Arguments are the actual values or data passed to the function when it is called. These values correspond to the parameters in the function definition. Parameters are the variables listed in the function definition that specify what kind of data the function will accept.

4.  What is the purpose of the `return` statement in a Python function?

ANS - Return statement sends a result back to the caller. This ends the function's execution immediately, and any code after it won't be executed.

5. What are iterators in Python and how do they differ from iterables?

ANS - An iterator is an object that keeps track of its current state during iteration and knows how to fetch the next item.
An iterable is any object in Python that can return an iterator, meaning it is capable of being looped over (iterated) using a for loop or any other iterator-based approach.

6.  Explain the concept of generators in Python and how they are defined.

ANS - In Python, generators are a special type of iterable, like lists or tuples, but instead of storing all their values in memory at once, they generate values on-the-fly as you loop over them. This makes generators particularly useful for working with large datasets or streams of data, where storing everything in memory would be inefficient.

7.  What are the advantages of using generators over regular functions?

ANS - The following are the advantages of generators over regular functions-

(i)More memory efficient
(ii)Efficient for large datasets
(iii)Can handle infinite sequences
(iv)More concise and readable code
(v)Remembers state between iterations
(vi)Can pause and resume execution
(vii)Fine control over iteration

8.  What is a lambda function in Python and when is it typically used?

ANS - A lambda function in Python is a small, anonymous function defined with the lambda keyword. It is used for defining simple functions in a concise way without needing to formally define them using the def keyword.

9. Explain the purpose and usage of the `map()` function in Python.

ANS - The main purpose of map() is to apply a transformation to each element in an iterable and return a new iterable containing the transformed elements. It is a very useful tool when you need to perform the same operation (such as a mathematical calculation, string manipulation, or data processing) on each element of a collection.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

ANS - map() applies a function to each item in an iterable.
reduce() reduces a sequence of items to a single value.
filter() filters elements of an iterable based on a condition.

```python
def word(a):
  return a[::-1]
word('hello world')
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power(expo):
    for i in range(expo + 1):
        yield 2 ** i
list(power(5))
```

 7. Implement a generator function that reads a file line by line and yields each line as a string.


```python
def read(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
```

 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.


```python
tuple_list = [(1, 3), (2, 1), (4, 2), (3, 4)]
sorted_list = sorted(tuple_list, key=lambda x: x[1])
print(sorted_list)

```

 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.


```python
l = [20, 30, 40, 50, 100]
list(map(lambda x: (x * 9/5) + 32, l))
```
